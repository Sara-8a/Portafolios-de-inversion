# Función de Métricas de Desempeño

```python
def metricas_desempeno(historia_port, tasa_rf, capital_inicial):

    # calcular rendimientos
    rendimientos = historia_port.pct_change().dropna()

    # rendimiento anualizado
    rendimiento_anual = rendimientos.mean()*252
    
    # rendimiento del mercado
    rendimiento_mercado = rendimiento_anual["benchmark"]

    # matriz de covarianza anualizada
    matriz_cov = rendimientos.cov()*252

    # dataframe donde guardaremos métricas
    tabla_metricas = pd.DataFrame(columns=historia_port.columns)

    # rendimiento promedio
    tabla_metricas.loc["Rendimiento Promedio"] = rendimiento_anual

    # rendimiento efectivo final
    tabla_metricas.loc["Rendimiento Efectivo"] = historia_port.iloc[-1]/capital_inicial - 1

    # volatilidad anual
    tabla_metricas.loc["Volatilidad"] = rendimientos.std()*np.sqrt(252)

    # sharpe
    tabla_metricas.loc["Sharpe"] = (tabla_metricas.loc["Rendimiento Promedio"]-tasa_rf)/tabla_metricas.loc["Volatilidad"]

    # downside
    retornos_neg = rendimientos.copy()
    retornos_neg[retornos_neg>0] = 0
    tabla_metricas.loc["Downside"] = retornos_neg.std()*np.sqrt(252)

    # upside
    retornos_pos = rendimientos.copy()
    retornos_pos[retornos_pos<0] = 0
    tabla_metricas.loc["Upside"] = retornos_pos.std()*np.sqrt(252)

    # omega
    tabla_metricas.loc["Omega"] = tabla_metricas.loc["Upside"]/tabla_metricas.loc["Downside"]

    # beta
    tabla_metricas.loc["Beta"] = matriz_cov["benchmark"]/matriz_cov.loc["benchmark","benchmark"]

    # alfa de jensen
    tabla_metricas.loc["Alpha Jensen"] = rendimiento_anual-(tasa_rf+tabla_metricas.loc["Beta"]*(rendimiento_mercado-tasa_rf))

    # treynor
    tabla_metricas.loc["Treynor"] = (rendimiento_anual-tasa_rf)/tabla_metricas.loc["Beta"]

    # sortino
    tabla_metricas.loc["Sortino"] = (rendimiento_anual-tasa_rf)/tabla_metricas.loc["Downside"]

    return tabla_metricas
```

# Estrategias op

### MIN VAR

```python
def pesos_min_var(rets):

    matriz_cov = rets.cov()
    n = len(rets.columns)

    f_obj = lambda w: w.T @ matriz_cov @ w

    w0 = np.ones(n)/n

    limites = [(0,1)]*n

    restr = lambda w: np.sum(w)-1

    resultado = minimize(f_obj, w0, bounds=limites,
                         constraints={"type":"eq","fun":restr})

    return resultado.x
```

### MAX SHARPE

```python
def pesos_max_sharpe(rets, rf):

    media = rets.mean()
    matriz_cov = rets.cov()

    n = len(rets.columns)

    sharpe = lambda w: -((w.T @ media - rf/252)/np.sqrt(w.T @ matriz_cov @ w))

    w0 = np.ones(n)/n

    limites = [(0,1)]*n

    restr = lambda w: np.sum(w)-1

    resultado = minimize(sharpe, w0, bounds=limites,
                         constraints={"type":"eq","fun":restr})

    return resultado.x
```

### SEMIVAR TARGET

```python
def pesos_semivar(rets, ret_bench):

    correl = rets.corr()

    dif = rets - ret_bench.values.reshape(-1,1)

    negativos = dif[dif<0].fillna(0)

    desv = np.array(negativos.std())

    matriz_semivar = desv.reshape(len(rets.columns),1) @ desv.reshape(1,len(rets.columns))*correl

    n = len(rets.columns)

    f = lambda w: w.T @ matriz_semivar @ w

    w0 = np.ones(n)/n

    limites = [(0.1,0.4)]*n

    restr = lambda w: np.sum(w)-1

    resultado = minimize(f, w0, bounds=limites,
                         constraints={"type":"eq","fun":restr})

    return resultado.x
```

### OPT OMEGA

```python
def pesos_omega(rets):

    n = len(rets.columns)

    perdidas = rets[rets<0].fillna(0)
    ganancias = rets[rets>0].fillna(0)

    downside = np.array(perdidas.std())
    upside = np.array(ganancias.std())

    omega = upside/downside

    f = lambda w: -np.sum(w*omega)

    w0 = np.ones(n)/n

    limites = [(0.1,0.4)]*n

    restr = lambda w: np.sum(w)-1

    resultado = minimize(f, w0, bounds=limites,
                         constraints={"type":"eq","fun":restr})

    return resultado.x
```

```python
rf = 0.0721

w_min = pesos_min_var(retornos_opt)

w_sharpe = pesos_max_sharpe(retornos_opt,rf)

w_semivar = pesos_semivar(retornos_opt,ret_ipc_opt)

w_omega = pesos_omega(retornos_opt)
```

## **Diccionario estrategias**

```python
estrategias = {

"Min Var": w_min,

"Max Sharpe": w_sharpe,

"Target Semivar": w_semivar,

"Omega": w_omega,

"benchmark":1
}
```

# **Funcion Backtesting**

```python
def backtesting_portafolio(precios, benchmark, estrategias, capital):

    rendimientos_act = precios.pct_change().dropna()

    rend_bench = benchmark.pct_change().dropna()

    historia = pd.DataFrame(index=rendimientos_act.index)

    for nombre in estrategias.keys():

        if nombre=="benchmark":

            historia[nombre] = capital*(1+rend_bench).cumprod()

        else:

            pesos = estrategias[nombre]

            ret_port = np.sum(rendimientos_act*pesos,axis=1)

            historia[nombre] = capital*(1+ret_port).cumprod()

    plt.figure(figsize=(12,6))

    plt.plot(historia)

    plt.title("Evolución de Estrategias de Inversión")

    plt.xlabel("Fecha")

    plt.ylabel("Valor del portafolio")

    plt.legend(historia.columns)

    plt.show()

    return historia
```

```python
capital = 1_000_000

historia = backtesting_portafolio(precios_bt,ipc_bt,estrategias,capital)
```
